```python
import pandas as pd

df = pd.read_csv('healthcare_data_cleaning_dataset.csv')
df.head()
```

```python
#Q1. Missing Data Identification
     #Scenario: 
     #The hospital suspects incomplete patient records.
     #Task:

# 1.Identify missing values in each column

# 2.Calculate percentage of missing data

#ANSWER:-

# Missing values in each column
missing_values = df.isnull().sum()

# Percentage of missing data
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Combine both results
missing_report = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage.round(2)
})

print(missing_report)
```

```python
#Q2. Handling Missing Age
     #Scenario:
     # Age is critical for medical analysis, but some values are missing.

     #Task:

     #Replace missing Age values with an appropriate method

     #Justify your choice (mean/median)
#ANSWER:-
import pandas as pd

df = pd.read_csv('healthcare_data_cleaning_dataset.csv')
df.head()

# Identify missing Age values
print("Missing Age values before:", df['Age'].isnull().sum())

# Replace missing Age values with Median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Check after replacement
print("Missing Age values after:", df['Age'].isnull().sum())
```

```python
#Q3. Handling Missing Treatment Cost
     #Scenario:
     #Treatment cost is highly skewed due to expensive treatments.

     #Task:

     #Handle missing Treatment_Cost values

     #Choose the correct imputation method and explain why

#ANSWER:-
import pandas as pd

df = pd.read_csv('healthcare_data_cleaning_dataset.csv')
df.head()

# Check missing values before
print("Missing Treatment_Cost before:", df['Treatment_Cost'].isnull().sum())

# Replace missing values with Median
df['Treatment_Cost'] = df['Treatment_Cost'].fillna(df['Treatment_Cost'].median())

# Check missing values after
print("Missing Treatment_Cost after:", df['Treatment_Cost'].isnull().sum())

#Choose the correct imputation method and explain why
#Explanation:-

#Median is the correct method because Treatment_Cost is highly skewed due to very expensive treatments.
#Mean can be heavily affected by extreme values (outliers), which may give an inaccurate result.
#Median is less sensitive to outliers and provides a more reliable value for replacing missing Treatment_Cost data.
```

```python
#Q4. Duplicate Patient Records
    #Scenario:
    #Some patient records were entered multiple times.
    #Task:

    #Identify duplicate rows

    #Remove duplicates

    #Compare dataset size before and after

#ANSWER:-

# Dataset size before removing duplicates
print("Dataset size before:", df.shape)

# Identify duplicate rows
duplicate_rows = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_rows)

# Remove duplicate rows
df = df.drop_duplicates()

# Dataset size after removing duplicates
print("Dataset size after:", df.shape)
```

```python
#Q5. Invalid Age Values (Data Quality Check)
     #Scenario:
     #Some patients have unrealistic age values (e.g., >100 or <0).
     #Task:

     #Detect such records

     #Decide whether to remove or correct them

#ANSWER:-


# Detect invalid age values (<0 or >100)
invalid_age = df[(df['Age'] < 0) | (df['Age'] > 100)]

# Display invalid records
print("Number of invalid age records:", len(invalid_age))

# Show invalid records
print(invalid_age)

# Remove invalid records
df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]

# Dataset size after removal
print("Dataset size after removing invalid ages:", df.shape)
```

```python
#Q6. Outlier Detection (Treatment Cost)
     #Scenario:
     #Extreme treatment costs are affecting analysis.

     #Task:

     #Detect outliers using IQR method

     #Display number of outliers

#ANSWER:-

# Calculate Q1 and Q3
Q1 = df['Treatment_Cost'].quantile(0.25)
Q3 = df['Treatment_Cost'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Treatment_Cost'] < lower_limit) | 
              (df['Treatment_Cost'] > upper_limit)]

# Display number of outliers
print("Number of outliers:", len(outliers))
```

```python
#Q7. Outlier Treatment
    #Scenario:
    #The business team wants to retain all records.

   #Task:

   #Apply capping (Winsorization) on Treatment_Cost

   #Use 5th and 95th percentile


#ANSWER:-

# Calculate 5th and 95th percentiles
lower_limit = df['Treatment_Cost'].quantile(0.05)
upper_limit = df['Treatment_Cost'].quantile(0.95)

# Apply capping (Winsorization)
df['Treatment_Cost'] = df['Treatment_Cost'].clip(lower=lower_limit,
                                                 upper=upper_limit)

# Display limits used
print("5th Percentile:", lower_limit)
print("95th Percentile:", upper_limit)

```

```python
#Q8. Transformation
     #Scenario:
     #Treatment cost is highly skewed.

     #Task:

     #Apply log transformation

     #Create a new column

     #Compare before vs after distribution

#ANSWER:-


import numpy as np

# Create a new column with log transformation
df['Log_Treatment_Cost'] = np.log1p(df['Treatment_Cost'])

# Compare before vs after distribution
print("Before Transformation:")
print(df['Treatment_Cost'].describe())

print("\nAfter Transformation:")
print(df['Log_Treatment_Cost'].describe())
```

```python
#Q9. Time-Based Missing Handling
     #Scenario:
     #Admission dates should follow a logical sequence.

     #Task:

     #Sort data by Admission_Date

     #Apply forward fill or backward fill where appropriate

     #Justify your choice


#ANSWER:-


# Convert Admission_Date to datetime
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

# Sort by Admission_Date
df = df.sort_values(by='Admission_Date')

# Apply Forward Fill
df['Admission_Date'] = df['Admission_Date'].ffill()

# Check missing values
print("Missing Admission_Date values:",
      df['Admission_Date'].isnull().sum())
```
